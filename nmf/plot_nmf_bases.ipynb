```python
import h5py
import matplotlib.pyplot as plt
import numpy as np

from neuropacks import ECOG
from nmf_plot_utils import plot_ecog_bases

%matplotlib inline
```

```python
data_path = '/Users/psachdeva/data/ECoG/data/r32_b7.mat'
grid_path = '/Users/psachdeva/data/ECoG/data/grdid.mat'
ecog = ECOG(
    data_path=data_path,
    grid_path=grid_path
)
```

```python
bases = h5py.File('fits/ecog_nmf.h5', 'r')
```

```python
uoi = bases['uoi'][:]
nmf = bases['nmf'][:]
```

```python
fig, axes = plt.subplots(3, 3, figsize=(16, 9))

plot_ecog_bases(components=uoi, ecog=ecog, fax=(fig, axes))

fig.suptitle(
    r'\textbf{UoI}$_{\textbf{NMF}}$ \textbf{Bases}',
    fontsize=30
)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('uoi_nmf_ecog_min.pdf', bbox_inches='tight')
```

```python
fig, axes = plt.subplots(3, 3, figsize=(16, 9))

plot_ecog_bases(components=nmf, ecog=ecog, fax=(fig, axes))

fig.suptitle(
    r'\textbf{NMF Bases}',
    fontsize=30
)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('nmf_ecog_min.pdf', bbox_inches='tight')
```
