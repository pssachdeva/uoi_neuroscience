```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from coupling_plot_utils import *
%matplotlib inline
```

# Indy 20160411_01

```python
indy_path = 'fits/nhp/indy_20160411_01.h5'
```

```python
axes = plot_coupling_grid(fits_path=indy_path)

# additional customizations #

# first row: selection ratio
for idx in range(3):
    axes[0, idx].set_xlim([0, 1])
    axes[0, idx].set_ylim([0, 1])
    axes[0, idx].plot([0, 1], [0, 1], color='gray', linewidth=3, zorder=-1)
    axes[0, idx].set_aspect('equal')

    axes[0, idx].set_xticks([0, 1])
    axes[0, idx].set_yticks([])

axes[0, 0].set_yticks([0, 1])

# second row: explained variance
for idx in range(3):
    axes[1, idx].set_xlim([0, 1])
    axes[1, idx].set_ylim([0, 1])
    axes[1, idx].plot([0, 1], [0, 1], color='gray', linewidth=3, zorder=-1)
    axes[1, idx].set_aspect('equal')

    axes[1, idx].set_xticks([0, 1])
    axes[1, idx].set_yticks([])

axes[1, 0].set_yticks([0, 1])

# third row: AIC
for idx in range(3):
    axes[2, idx].set_xlim([-3000, 500])
    axes[2, idx].set_ylim([-3000, 500])
    axes[2, idx].plot([-3000, 500], [-3000, 500], color='gray', linewidth=3, zorder=-1)
    axes[2, idx].set_aspect('equal')

    axes[2, idx].set_xticks([])
    axes[2, idx].set_yticks([])

# fourth row: AIC
for idx in range(3):
    axes[3, idx].set_xlim([-3000, 500])
    axes[3, idx].set_ylim([-3000, 500])
    axes[3, idx].plot([-3000, 500], [-3000, 500], color='gray', linewidth=3, zorder=-1)
    axes[3, idx].set_aspect('equal')

    axes[3, idx].set_xticks([])
    axes[3, idx].set_yticks([])

# row labels
axes[0, 0].set_title(r'\textbf{UoI, $R^2$}', fontsize=21)
axes[0, 1].set_title(r'\textbf{UoI, AIC}', fontsize=21)
axes[0, 2].set_title(r'\textbf{UoI, BIC}', fontsize=21)

# column labels
axes[0, 0].set_ylabel(
    r'\textbf{Selection}' '\n' r'\textbf{Ratio}', fontsize=21
)
axes[1, 0].set_ylabel(
    r'\textbf{Explained}' '\n' r'\textbf{Variance}', fontsize=21
)
axes[2, 0].set_ylabel(r'\textbf{AIC}', fontsize=21)
axes[3, 0].set_ylabel(r'\textbf{BIC}', fontsize=21)

# overall title
plt.gcf().text(
    x=0.55, y=1.03,
    s=r'\textbf{M1, Indy (04/11/2016--01)}',
    horizontalalignment='center',
    verticalalignment='center',
    fontsize=30
)

plt.tight_layout(h_pad=0., w_pad=0.)
#plt.savefig('indy_20160411_01.pdf', bbox_inches='tight')
```
